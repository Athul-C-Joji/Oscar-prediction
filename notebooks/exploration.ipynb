```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better-looking plots
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

print("‚úÖ Libraries imported successfully!")
```

```python
# Load the processed data
df = pd.read_csv('../data/processed/best_picture_clean.csv')

print(f"Dataset shape: {df.shape}")
print(f"\nColumns: {list(df.columns)}")
print(f"\nFirst 10 rows:")
df.head(10)
```

```python
# First, let's see what columns we actually have
print("üìä Dataset Info:")
print(f"Shape: {df.shape}")
print(f"\nColumn names:")
print(df.columns.tolist())
print("\nüîç First few rows:")
print(df.head())
```

```python
# Visualize Best Picture winners over time
winners_df = df[df['winner'] == 1]

plt.figure(figsize=(14, 6))
plt.scatter(winners_df['year_ceremony'], winners_df['film'], s=100, alpha=0.6, color='gold', edgecolors='black', linewidth=1.5)
plt.title('Oscar Best Picture Winners by Year', fontsize=16, fontweight='bold')
plt.xlabel('Ceremony Year', fontsize=12)
plt.ylabel('Film', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3, axis='x')
plt.tight_layout()
plt.show()

print(f"\nüèÜ Total Best Picture winners in dataset: {len(winners_df)}")
print(f"üìÖ Year range: {winners_df['year_ceremony'].min()} - {winners_df['year_ceremony'].max()}")
```

```python
# Analyze relationship between total nominations and winning
plt.figure(figsize=(12, 6))

# Separate winners and non-winners
winners = df[df['winner'] == 1]
losers = df[df['winner'] == 0]

# Create scatter plot
plt.scatter(losers['total_nominations'], losers['nomination_share'], 
           alpha=0.5, s=50, label='Did Not Win', color='lightblue', edgecolors='black', linewidth=0.5)
plt.scatter(winners['total_nominations'], winners['nomination_share'], 
           alpha=0.8, s=150, label='Winner', color='gold', edgecolors='black', linewidth=2, marker='*')

plt.title('Total Nominations vs Nomination Share (Winners vs Non-Winners)', fontsize=14, fontweight='bold')
plt.xlabel('Total Nominations', fontsize=12)
plt.ylabel('Nomination Share', fontsize=12)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\nüìä Average nominations for winners: {winners['total_nominations'].mean():.2f}")
print(f"üìä Average nominations for non-winners: {losers['total_nominations'].mean():.2f}")
```
